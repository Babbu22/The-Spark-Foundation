### Cleaning The Data

```python
#Delete uneccessary column
df = df.drop(['Country'], axis = 1)
```

```python
#creating a scatter plot
df.plot.scatter(x='Discount',y='Profit',c='blue',s=50,figsize=(12,5))
plt.show()
```

```python
#All data in code
sns.set_palette('viridis')
sns.pairplot(df)
plt.show()
```

```python
#category vs sales bar graph
plt.bar('Category','Sales',data=df)
plt.title('Sales vs Category',size=15)
plt.xlabel('Category',size=13)
plt.ylabel('Sales',size=13)
plt.show()
```

```python
#shipping models
plt.figure(figsize=(6,5))
sns.countplot('Ship Mode',data=df,palette='Blues_d')
plt.title('Popular Shipping Mode',size=14)
plt.xlabel('\n Shipping Mode',size=12)
plt.ylabel('Number of orders',size=12)
plt.xticks(fontsize=12)
plt.show()
```

```python
#Quantities ordered by each segment
dl_group = df.groupby('Segment')['Quantity'].sum().reset_index()
print(dl_group)
labels=df['Segment'].unique()
colors=['lightskyblue','pink','maroon']
plt.figure(figsize=(5,5))
plt.pie(dl_group['Quantity'],autopct='%1.1f%%',labels=labels,explode=(0.02,0.02,0.02),colors=colors)
plt.title('Quantities ordered by each segment',size=15)
plt.show()
```

###### Shiping Mode vs Count

```python
sns.catplot('Ship Mode',data=df,hue='Segment',kind='count',palette='Blues_d',aspect=1.5,height=6)
plt.show()
```

###### City each of least and most quantities ordered

```python
sns.catplot('State',kind='count',data=df,palette='ocean',height=6,aspect=1.5)
plt.xticks(rotation=90)
plt.show()
```

###### Quantites ordered region wise

```python
dl_region=df.groupby('Region')['Quantity'].sum().reset_index()
print(dl_region)
labels = dl_region['Region'].unique()
plt.figure(figsize=(5,5))
plt.pie(dl_region['Quantity'],autopct='%1.1f%%',explode=(0.02,0.02,0.02,0.02),)
plt.title('Quantities ordered by each region', size=13)
plt.show()
```

```python
#Highest Selling categories
dl_cats = df.groupby('Category')['Quantity'].sum().reset_index()
dl_cats
```

```python
#Most profitable products
dltop10_items =df.groupby('Sub-Category')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False)
dltop10_items.reset_index(drop=True,inplace=True)
dltop10_items=dltop10_items.head(10)
dltop10_items
```

```python
#visualizung the top 10 profitable products
sns.catplot('Sub-Category','Profit',data=dltop10_items,kind='bar',aspect=1.5,height=9,palette='ch:2.5,-.2,dark=.3')
plt.title('Top 10 profitable products',size=20)
plt.xticks(size=15)
plt.yticks(size=15)
plt.ylabel('Cumulative profit',size=18)
plt.xlabel('Products',size=18)
plt.show()
```

```python
#Top Profitable city
dltop10_items = df.groupby('City')['Profit'].sum().reset_index().sort_values(by='Profit',ascending =False)
dltop10_items = dltop10_items.head(10)
dltop10_items
```

```python
#Top 10 profitable cities
sns.catplot('City','Profit',data=dltop10_items,kind='bar',aspect=2,height=8,palette='RdBu')
plt.title('Top 10 cities with profit inflow',size=25)
plt.xticks(size=15)
plt.yticks(size=15)
plt.ylabel('Cumulate profit',size=18)
plt.xlabel('City',size=18)
plt.show()
```
